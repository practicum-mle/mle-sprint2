### Признаки

```python
fig, axs = plt.subplots(2, 2)
fig.tight_layout(pad=1.6)
fig.set_size_inches(16.5, 12.5, forward=True)


x = "balcony"
y = "building_series_id"
stat = ["count"]
agg_df = pd.DataFrame(df.groupby(x).agg(stat)[y]).reset_index()
sns.barplot(
    data=agg_df,
    x=x,
    y=stat[0],
    ax=axs[0, 0]
)
axs[0, 0].set_title(f'Count {y} by {x} in train dataframe')

x = "building_type_int"
y = "building_series_id"
stat = ["count"]
agg_df = pd.DataFrame(df.groupby(x).agg(stat)[y]).reset_index()
sns.barplot(
    data=agg_df,
    x=x,
    y=stat[0],
    ax=axs[1, 0]
)
axs[1, 0].set_title(f'Count {y} by {x} in train dataframe')
axs[1, 0].set_xticklabels(df[x].unique(), rotation = 45);

x = "renovation"
y = "building_series_id"
stat = ["count"]
agg_df = pd.DataFrame(df.groupby(x).agg(stat)[y]).reset_index()
sns.barplot(
    data=agg_df,
    x=x,
    y=stat[0],
    ax=axs[0, 1]
)
axs[0, 1].set_title(f'Count {y} by {x} in train dataframe')


x = "rooms"
y = "building_series_id"
stat = ["count"]
agg_df = pd.DataFrame(df.groupby(x).agg(stat)[y]).reset_index()
sns.barplot(
    data=agg_df,
    x=x,
    y=stat[0],
    ax=axs[1, 1]
)
axs[1, 1].set_title(f'Count {y} by {x} in train dataframe')


plt.savefig(os.path.join("assets", "cat_features_1"))
```

```python
x = "building_series_id"
binary_columns = [
    "is_apartment", 
    "studio", 
]
stat = ["count"]
```

```python
df.groupby(binary_columns).agg(stat[0])[x].reset_index().sort_values(
    by=x, ascending=False
).head(10)
```

```python
agg_binary = df[binary_columns].apply(pd.Series.value_counts).T.reset_index()
agg_binary_melted = agg_binary.melt(id_vars='index', var_name='Value', value_name='Counts')

sns.barplot(data=agg_binary_melted, x='index', y='Counts', hue='Value')
plt.title(f"Binary columns distribution")
plt.xticks(rotation=45);

plt.savefig(os.path.join("assets", "cat_features_2_binary"))
```

```python
x = "build_year"
charges_columns = [
    "living_area",
    "last_price",
]
stats = ["mean", "median", lambda x: x.mode().iloc[0]]
```

```python
df[charges_columns[0]] = df[charges_columns[0]].astype(float).astype(int)
df[charges_columns[1]] = df[charges_columns[1]].astype(float).astype(int)
```

```python
charges_monthly_agg = df[[x] + [charges_columns[1]]].groupby([x]).agg(stats).reset_index()
charges_monthly_agg.columns = charges_monthly_agg.columns.droplevel()
charges_monthly_agg.columns = [x, "living_area_mean", "living_area_median", "living_area_mode"]

charges_total_agg = df[[x] + [charges_columns[0]]].groupby([x]).agg(stats).reset_index()
charges_total_agg.columns = charges_total_agg.columns.droplevel()
charges_total_agg.columns = [x, "last_price_mean", "last_price_median", "last_price_mode"]

charges_monthly_agg = charges_monthly_agg[charges_monthly_agg["build_year"] > 1950].sort_values("build_year")
charges_total_agg = charges_total_agg[charges_total_agg["build_year"] > 1950].sort_values("build_year")
```

```python
fig, axs = plt.subplots(2, 1)
fig.tight_layout(pad=2.5)
fig.set_size_inches(6.5, 5.5, forward=True)


sns.lineplot(charges_monthly_agg, ax=axs[0], x=x, y="living_area_mean")
sns.lineplot(charges_monthly_agg, ax=axs[0], x=x, y="living_area_median")
sns.lineplot(charges_monthly_agg, ax=axs[0], x=x, y="living_area_mode")
axs[0].set_title(f"Statistics for {charges_columns[0]} by {x}")

sns.lineplot(charges_total_agg, ax=axs[1], x=x, y="last_price_mean")
sns.lineplot(charges_total_agg, ax=axs[1], x=x, y="last_price_median")
sns.lineplot(charges_total_agg, ax=axs[1], x=x, y="last_price_mode")
axs[1].set_title(f"Statistics for {charges_columns[1]} by {x}")

# plt.legend(['mean', 'median', 'mode'])
# plt.savefig(os.path.join(ASSETS_DIR, "charges_by_date"))
```
